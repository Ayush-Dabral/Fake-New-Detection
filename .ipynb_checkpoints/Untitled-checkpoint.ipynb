```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
import re
import string
```

```python
fake_text = pd.read_csv('Fake.csv')
true_text = pd.read_csv('True.csv')
```

```python
fake_text.head()
```

```python
true_text.head()
```

```python
fake_text["label"] = 0
true_text["label"] = 1
```

```python
fake_text.shape, true_text.shape
```

```python
merged_text = pd.concat([fake_text, true_text], axis=0)
```

```python
merged_text.head()
```

```python
df = merged_text.copy()
df = df.drop(['subject', 'date'], axis=1)
df.head()
```

# Dataset Expansion

```python
nds = pd.read_csv('New dataset.csv')
```

```python
df = pd.concat([df, nds], axis=0)
```

```python
df = df.drop(['Unnamed: 0'], axis = 1)
df = df.dropna()
df.head()
```

```python
newtext = df.sample(frac=1)
y = newtext['label']
newtext = newtext.drop('label', axis=1)
```

```python
newtext.head()
```

```python
newtext.shape, y.shape
```

```python
newtext.reset_index(inplace=True)
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
corpus = []
for i in range(0, len(newtext)):
        review = re.sub('[^a-zA-Z]', ' ', newtext['title'][i])
        review = re.sub('[^a-zA-Z]', ' ', newtext['text'][i])
        review = review.lower()
        review = review.split()
        review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
        review = ' '.join(review)
        corpus.append(review)
```

```python
corpus
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
import sys,os
os.chdir(sys.path[0])
stopwords = STOPWORDS
wc = WordCloud(background_color = 'black', stopwords = stopwords, height = 1000, width = 2000)
wc.generate(str(corpus))
wc.to_file('wordcloud_output.png')
plt.figure( figsize=(15,7) )
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,ngram_range=(1,3))
x = cv.fit_transform(corpus).toarray()
x.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

# Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(criterion='entropy')
dtc.fit(x_train, y_train)
y_preddtc = dtc.predict(x_test)
```

```python
import mlxtend
from sklearn import metrics
from mlxtend.plotting import plot_confusion_matrix
from sklearn.metrics import confusion_matrix 
scoredtc = metrics.accuracy_score(y_test, y_preddtc)*100
print("accuracy : %f" %scoredtc)
cm = metrics.confusion_matrix(y_test, y_preddtc, labels=[0,1])
fig, ax = plot_confusion_matrix(conf_mat = confusion_matrix(y_test, y_preddtc), show_absolute = True, show_normed = True, colorbar = True)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(solver='lbfgs', max_iter=50000)
lr.fit(x_train, y_train)
y_predlr = lr.predict(x_test)
```

```python
scorelr = metrics.accuracy_score(y_test, y_predlr)*100
print("accuracy : %f" %scorelr)
cm = metrics.confusion_matrix(y_test, y_predlr, labels=[0,1])
fig, ax = plot_confusion_matrix(conf_mat = confusion_matrix(y_test, y_predlr), show_absolute = True, show_normed = True, colorbar = True)
```

# Support Vector Machine

```python
from sklearn.svm import LinearSVC
svm = LinearSVC(max_iter=5000000)
svm.fit(x_train, y_train)
y_predsvm = svm.predict(x_test)
```

```python
scoresvm = metrics.accuracy_score(y_test, y_predsvm)*100
print("accuracy : %f" %scoresvm)
cm = metrics.confusion_matrix(y_test, y_predsvm, labels=[0,1])
fig, ax = plot_confusion_matrix(conf_mat = confusion_matrix(y_test, y_predsvm), show_absolute = True, show_normed = True, colorbar = True)
```

# Gradient Boosting Classifier

```python
from sklearn.ensemble import GradientBoostingClassifier
gbc = GradientBoostingClassifier(random_state = 0)
gbc.fit(x_train, y_train)
y_predgbc = gbc.predict(x_test)
```

```python
scoregbc = metrics.accuracy_score(y_test, y_predgbc)*100
print("accuracy : %f" %scoregbc)
cm = metrics.confusion_matrix(y_test, y_predgbc, labels=[0,1])
fig, ax = plot_confusion_matrix(conf_mat = confusion_matrix(y_test, y_predgbc), show_absolute = True, show_normed = True, colorbar = True)
```

# Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(random_state = 0)
rfc.fit(x_train, y_train)
y_predrfc = rfc.predict(x_test)
```

```python
scorerfc = metrics.accuracy_score(y_test, y_predrfc)*100
print("accuracy : %f" %scorerfc)
cm = metrics.confusion_matrix(y_test, y_predrfc, labels=[0,1])
fig, ax = plot_confusion_matrix(conf_mat = confusion_matrix(y_test, y_predrfc), show_absolute = True, show_normed = True, colorbar = True)
```

# The most accurate

```python
import pickle
```

```python
if scoredtc>scorelr and scoredtc>scoresvm and scoredtc>scoregbc and scoredtc>scorerfc :
    print("The most accurate is : Decision Tree Classifier for this dataset.")
    with open('model_pickle','wb') as f:
        pickle.dump((dtc,cv),f)
elif scorelr>scoredtc and scorelr>scoresvm and scorelr>scoregbc and scorelr>scorerfc :
    print("The most accurate is : Logistic Regression for this dataset.")
    with open('model_pickle','wb') as f:
        pickle.dump((lr,cv),f)
elif scoresvm>scoredtc and scoresvm>scorelr and scoresvm>scoregbc and scoresvm>scorerfc :
    print("The most accurate is : Support Vector Classifier for this dataset.")
    with open('model_pickle','wb') as f:
        pickle.dump((svm,cv),f)
elif scoregbc>scoredtc and scoregbc>scoresvm and scoregbc>scorelr and scoregbc>scorerfc :
    print("The most accurate is : Gradient Boosting Classifier for this dataset.")
    with open('model_pickle','wb') as f:
        pickle.dump((gbc,cv),f)
else :
    print("The most accurate is : Random Forest Classifier for this dataset.")
    with open('model_pickle','wb') as f:
        pickle.dump((rfc,cv),f)
```
